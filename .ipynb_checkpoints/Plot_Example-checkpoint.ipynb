## Plotting Example Script

### Loading packages

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline 

import warnings
from copy import deepcopy
import numpy as np
import pickle

# import for plotting
import os
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
plt.style.use('ggplot')
```

Pre-define a dummy class

```python
warnings.filterwarnings('ignore')
```

```python
class dummy:
    pass
```

### Begin plot


Running the algorithm will generate a folder named "result**", and we can see inside:

+ Settings.text: the description of the model settings.
+ bnnacc/plot_bnn: if running BNN.
+ snnacc/plot_snn: if running SNN.

If running in parallel, results of all nodes will be saved, named "bnnacc0, bnnacc1, bnnacc2" for example.

In order to simply plot the results, I saved results of all experiments in folder 'results', and we can find 'README_cifar.xlsx' documenting all model settings.

### Functions to read results

Since I'm not running in parallel for the very early experiments, the output is slightly different. So I write two different functions to read output using horovod or not.

+ genBnnlist_h/genSnnlist_h: read the results of BNN/SNN using parallel running ('h' means 'horovod').
+ genBnnlist/genSnnlist: read the results of BNN/SNN not using parallel running.

In "README_cifar.xlsx", I'm using horovod after experiment ID '49', since epochs=20h, and 'h' means 'horovod'.

```python
def genBnnlist_h(bnn_trial_list, num_nodes):
    bnnlist = []
    for x in bnn_trial_list:
        makedir = "results/result"+str(x)+"/"
        baynet = dummy()
        baynet.runtime = 0.0
        devAccList = []
        accList = []
        for i in range(num_nodes):
            with open(makedir + "plot_bnn" + str(i),"rb") as load:
                niter, runtime, _, devAcc = pickle.load(load)
            with open(makedir + "bnnacc" + str(i),"rb") as load:
                acc = pickle.load(load)
            baynet.runtime += np.array(runtime) / num_nodes
            decAcc = np.array(devAcc)
            if len(decAcc.shape) == 1:
                decAcc = decAcc.reshape((decAcc.shape[0],1))
            devAccList.append(decAcc)
            accList.append(acc)
        
        baynet.niter = np.array(niter) * num_nodes
        baynet.devAcc = np.concatenate(devAccList, axis=1)
        baynet.acc = np.concatenate(accList)
        meanval = baynet.devAcc.mean(axis=1)
        stdval = baynet.devAcc.std(axis=1)
        baynet.devAccMean = meanval
        baynet.devAccUp = meanval + stdval
        baynet.devAccDown = meanval - stdval
    
        bnnlist.append(baynet)
    return bnnlist

def genBnnlist(bnn_trial_list):
    bnnlist = []
    for x in bnn_trial_list:
        makedir = "results/result"+str(x)+"/"

        baynet = dummy()
        with open(makedir+"plot_bnn","rb") as load:
            baynet.niter, baynet.runtime, baynet.loss, baynet.devAccMean, baynet.devAccUp, baynet.devAccDown = pickle.load(load)

        with open(makedir+"bnnacc","rb") as load:
            baynet.acc = pickle.load(load)

        bnnlist.append(baynet)
    return bnnlist

def genSnnlist_h(snn_trial_list, num_nodes):
    snnlist = []
    for x in snn_trial_list:
        makedir = "results/result"+str(x)+"/"

        net = dummy()
        with open(makedir+"plot_snn","rb") as load:
            niter, net.runtime, net.loss, net.devAcc = pickle.load(load)

        with open(makedir+"snnacc" + str(0),"rb") as load:
            net.acc = pickle.load(load)
        net.niter = np.array(niter) * num_nodes

        snnlist.append(net)
    return snnlist

def genSnnlist(snn_trial_list):
    snnlist = []
    for x in snn_trial_list:
        makedir = "results/result"+str(x)+"/"

        net = dummy()
        with open(makedir+"plot_snn","rb") as load:
            net.niter, net.runtime, net.loss, net.devAcc = pickle.load(load)

        with open(makedir+"snnacc","rb") as load:
            net.acc = pickle.load(load)

        snnlist.append(net)
    return snnlist
```

#### Example 1: no parallel running, only fully connected layer is random.

```python
bnn_trial_list = [7,8,10,11,13,14,9.5]
bnn_trial_list = list(map(lambda x:str(x)+"c",bnn_trial_list))
snn_trial_list = ['9c']

[net] = genSnnlist(snn_trial_list)
[baynet0, baynet1, baynetp5, baynetp1, baynetp2, baynetp3, baynetp4] = genBnnlist(bnn_trial_list)
```

```python
fig = plt.figure(figsize=(8,6))
ax1 = fig.add_subplot(111)
ax1.plot(net.niter,net.devAcc,label="NN",color="b", linestyle=":")
ax1.fill_between(baynet0.niter, baynet0.devAccUp, baynet0.devAccDown, label="BNN scale 0 fill")
plt.fill_between(baynet0.niter, baynetp1.devAccUp, baynetp1.devAccDown, label="BNN scale 0.1 fill")
plt.fill_between(baynet0.niter, baynetp2.devAccUp, baynetp2.devAccDown,  label="BNN scale 0.2 fill")
plt.fill_between(baynet0.niter, baynetp3.devAccUp, baynetp3.devAccDown,  label="BNN scale 0.3 fill")
plt.fill_between(baynet0.niter, baynetp5.devAccUp, baynetp5.devAccDown,  label="BNN scale 0.5 fill")
plt.fill_between(baynet0.niter, baynet1.devAccUp, baynet1.devAccDown,  label="BNN scale 1 fill")

ax1.set_xlabel("iterations")
ax1.set_ylabel("accuracy")
ax1.set_xlim(-500,7000)
ax2 = ax1.twiny()
ax2.set_xlabel("epochs")
ax2.set_xlim(-500,7000)
ax2.set_xticks(np.array([1,2,3])*6700/3)
ax2.set_xticklabels(['5th','10th','15th'])
ax2.axvline(x=6700/3,linestyle=":",color='black')
ax2.axvline(x=6700/3*2,linestyle=":",color='black')
ax2.axvline(x=6700/3*3,linestyle=":",color='black')
ax1.legend(bbox_to_anchor=(1.05, 0.5), loc=2, borderaxespad=0.)
plt.show()
```

#### Example 4: using horovod

```python
bnn_trial_list_1h = ['157c']
bnn_trial_list_2h = ['158c']
bnn_trial_list_4h = ['159c1', '159c2']
bnn_trial_list_8h = ['160c']
bnn_list_16h = ['node16bnn1','node16bnn2']
bnn_list_32h = ['node32bnn']

[baynet1h] = genBnnlist_h(bnn_trial_list_1h, num_nodes=1)
[baynet2h] = genBnnlist_h(bnn_trial_list_2h, num_nodes=2)
[baynet4h1, baynet4h2] = genBnnlist_h(bnn_trial_list_4h, num_nodes=4)
[baynet8h] = genBnnlist_h(bnn_trial_list_8h, num_nodes=8)
[baynet16h1, baynet16h2] = genBnnlist_h(bnn_list_16h, num_nodes=16)
[baynet32h] = genBnnlist_h(bnn_list_32h, num_nodes=32)
```

```python
snn_trial_list_1h = ['147c']
snn_trial_list_2h = ['148c']
snn_trial_list_4h = ['149c1', '149c2']
snn_trial_list_8h = ['150c']
snn_list_16h = ['node16snn1','node16snn2']
snn_list_32h = ['node32snn']

[net1h] = genSnnlist_h(snn_trial_list_1h, num_nodes=1)
[net2h] = genSnnlist_h(snn_trial_list_2h, num_nodes=2)
[net4h1, net4h2] = genSnnlist_h(snn_trial_list_4h, num_nodes=4)
[net8h] = genSnnlist_h(snn_trial_list_8h, num_nodes=8)
[net16h1, net16h2] = genSnnlist_h(snn_list_16h, num_nodes=16)
[net32h] = genSnnlist_h(snn_list_32h, num_nodes=32)
```

```python
# node 16
snn_list_16h = ['node16snn1','node16snn2']
bnn_list_16h = ['node16bnn1','node16bnn2']
[baynet16h1, baynet16h2] = genBnnlist_h(bnn_list_16h, num_nodes=16)
[net16h1, net16h2] = genSnnlist_h(snn_list_16h, num_nodes=16)
```

```python
df = {"1-node":[baynet1h.runtime.mean(),np.mean(net1h.runtime)],
      "2-node":[baynet2h.runtime.mean(),np.mean(net2h.runtime)],
      "4-node":[(baynet4h1.runtime+baynet4h2.runtime).mean()/2, (np.array(net4h1.runtime)+np.array(net4h2.runtime)).mean()/2],
      "8-node":[baynet8h.runtime.mean(),np.mean(net8h.runtime)],
      "16-node":[(baynet16h1.runtime+baynet16h2.runtime).mean()/2, (np.array(net16h1.runtime)+np.array(net16h2.runtime)).mean()/2],
      "32-node":[baynet32h.runtime.mean(),np.mean(net32h.runtime)]
     }
```
